A dispersion relation from S.Berg and F.Ruggiero:
 $$1 = - \Delta Q_{\rm coh}\int_0^{\infty}dJ_x\int^{\infty}_0 dJ_y \dfrac{J_x \dfrac{\partial\psi(J_x, J_y)}{\partial J_x}}{Q-Q_x(J_x, J_y)}$$

## Octupoles

 For octupole magnets the incoherent betatron tune spread is linear with amplitude:
\begin{align*}
    &\Delta Q_{x} = a_{xx}J_x / \epsilon_{x} + a_{yx} J_y /\epsilon_{x}, \\
    &a_{xx} \propto I_{\rm oct}\epsilon_{\rm n}/\gamma^2
\end{align*}
where $I_{\rm oct}$ is octupole current, $\epsilon_{\rm n}$ -- normalized emittance, $\gamma$ -- relativistic gamma.

## SIS18 electron cooler as an electron lens

```python
from scipy.constants import m_e, c, e, m_p, epsilon_0
import numpy as np
epsn_x = 12.5e-6
beta_x = 8
I_e = 1.5
U_e = 7e3
d_cath = 0.0254
L_e = 2.8
gamma_e = 1 + U_e * e / (m_e * c**2)
beta_e = np.sqrt(1-gamma_e**-2)
I_a = 17e3
```

```python
def max_tune_shift(Ekin, f_e, A=1, Z=1):
    gamma = 1 + Ekin * e / (m_p * c**2)
    beta = np.sqrt(1-gamma**-2)
    sigma_x = np.sqrt(epsn_x*beta_x/gamma)
    sigma_e = 0.0254/8*np.sqrt(f_e)
    dQmax = Z/A*I_e/I_a*m_e/m_p*(sigma_x/sigma_e)**2*L_e/(4*np.pi*epsn_x)*(1+beta_e*beta)/(beta_e*beta)
    return dQmax
def tune_shift_sc(Ekin, intensity, A=1, Z=1):
    gamma = 1 + Ekin * e / (m_p * c**2)
    beta = np.sqrt(1-gamma**-2)
    beta_x = 10
    tune_x = 4.2
    R = 216.72/(2*np.pi)
    sigma_x = np.sqrt(epsn_x*beta_x/gamma)
    r_0 = e**2/(4*np.pi*epsilon_0*m_p*c**2)
    dQmax = -Z**2/A*intensity*r_0*R/(4*np.pi*gamma**3*beta**2*tune_x)*1/sigma_x**2
    return dQmax
```

```python
from matplotlib import pyplot as plt
def get_beta(x):
    return np.sqrt(1-(1+x*e/(m_p*c**2))**-2)
plt.plot(np.linspace(11.4e6, 200e6, 1000), get_beta(np.linspace(11.4e6, 200e6, 1000)))
plt.axhline(beta_e)
plt.show()
```

```python
print('For protons at top energy in SIS18:')
print('Tune shift from an electron lens is: {0:.4f}'.format(max_tune_shift(4.5e9, 4)))
print('Tune shift from space-charge is: {0:.6f}'.format(tune_shift_sc(4.5e9, 5e10)))
print('For protons at injection energy in SIS18:')
print('Tune shift from an electron lens is: {0:.4f}'.format(max_tune_shift(11.4e6, 8)))
print('Tune shift from space-charge is: {0:.4f}'.format(tune_shift_sc(11.4e6, 5e10)))
```

## SIS100 Octupoles



```python
import numpy as np
from scipy.constants import e, c, m_p
def get_detuning_from_SIS100_octupoles(Ekin, K3):
    beta_xF = 5.8 
    beta_xD = 17.2
    beta_yF = 17.7
    beta_yD = 5.8
    Lm = 0.75
    Nf = 6
    Nd = -6
    A, Q = 238, 28
    Brho = 3.3357*A/Q*Ekin/1e9
    print(Brho)
    axx = 3/(8*np.pi)*(e/c*Ekin)*Lm*(Nf*beta_xF**2*K3/6+2*Nd*beta_xD**2*K3/6)/(e/c*Ekin)
    axy = 3/(8*np.pi)*(e/c*Ekin)*Lm*(2*Nf*beta_xF*beta_yF*K3/6+2*Nd*beta_xD*beta_yD*K3/6)/(e/c*Ekin)
    ayx = 3/(8*np.pi)*(e/c*Ekin)*Lm*(2*Nf*beta_xF*beta_yF*K3/6+2*Nd*beta_xD*beta_yD*K3/6)/(e/c*Ekin)
    ayy = 3/(8*np.pi)*(e/c*Ekin)*Lm*(Nf*beta_yF**2*K3/6+2*Nd*beta_yD**2*K3/6)/(e/c*Ekin)
    return ((axx, axy), (ayx, ayy))
K3 = 50
Ekin = 2e8
epsn_x = 6e-6
epsn_y = 2.5e-6
gamma = 1 + Ekin * e / (m_p * c**2)
a = (get_detuning_from_SIS100_octupoles(Ekin, K3))
b = np.array((epsn_x/gamma, epsn_y/gamma))
((axx, axy), (ayx, ayy)) = a*b
print(((axx, axy), (ayx, ayy)))
```
